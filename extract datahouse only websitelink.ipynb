```
import urllib
from bs4 import BeautifulSoup
from urllib.request import urlopen, Request
import time
```

```
#headers ต้องใส่แบบนี้ทุกครั้งจะได้ไม่มีปัญหาในการดึงข้อมูล
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.3'}

#QuotePageList แต่ล่ะ link คือแต่ล่ะหน้า website ที่เราต้องการดึงข้อมูลอันนี้ต้องทำด้วยมือเองว่าจะดึงข้อมูลจากหน้าไหนบ้าน
QuotePageList = [
'https://baania.com/en/projects?province=3781&ptype=1',
'https://baania.com/en/projects?page=1&province=3781&ptype=1',
'https://baania.com/en/projects?page=2&province=3781&ptype=1',
'https://baania.com/en/projects?page=3&province=3781&ptype=1',
'https://baania.com/en/projects?page=4&province=3781&ptype=1',
'https://baania.com/en/projects?page=5&province=3781&ptype=1',
'https://baania.com/en/projects?page=6&province=3781&ptype=1',
'https://baania.com/en/projects?page=7&province=3781&ptype=1',
'https://baania.com/en/projects?page=8&province=3781&ptype=1',
'https://baania.com/en/projects?page=9&province=3781&ptype=1',
'https://baania.com/en/projects?page=10&province=3781&ptype=1',
'https://baania.com/en/projects?page=11&province=3781&ptype=1',      
]

#HouseLinkList คือ list ที่สร้างเพิ่มขึ้นมาเพื่อเก็บข้อมูล link บ้านในแต่ล่ะหน้าทุกหน้าที่ดึงข้อมูลมา
HouseLinkList = [] # To collect link list
```

```
for QuotePage in QuotePageList:
    print('----------------- Current Page is -----------------')
    print(QuotePage)
    
    req = Request(url=QuotePage, headers=headers) 
    Page = urlopen(req).read() 
    Soup = BeautifulSoup(Page,'html.parser')
#MainTag คือ code ที่ลากครอบคลุมข้อมูลบ้านทุกหลังในหน้านั้นๆ
    MainTag = Soup.find('div', attrs={'class': '_1bUgSkfUE8fucpZ0CSkKXt container-fluid'})
#LinkTagList คือ ชื่อโครงการบ้านแต่ล่ะหลังที่มี url อยู่ด้านล่าง
    LinkTagList = MainTag.findAll('h5', attrs={'class': 'mb-2 text-truncate'})

    print('Number of house inside this page is = ' + str(len(LinkTagList)))


    for LinkTag in LinkTagList:
        Link = LinkTag.find('a').get('href') # อันนี้คือ keyword ของการดึง href จาก Tag
        HouseLinkList.append(Link)

    time.sleep(2)
    
print('Program Completed')
```
