```
import urllib
from bs4 import BeautifulSoup
from urllib.request import urlopen, Request
import time
import openpyxl
from google.colab import files
```

```
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.3'}
QuotePage = "https://baania.com/en/project/%E0%B8%9E%E0%B8%B2%E0%B8%A3%E0%B9%8C%E0%B8%84%E0%B9%80%E0%B8%A7%E0%B8%A2%E0%B9%8C-%E0%B9%81%E0%B8%AD%E0%B8%97%E0%B8%AD%E0%B8%B5%E0%B8%8B?s=search&pid=BKK2578&province=3781&district=3936&ptype=1"

req = Request(url=QuotePage, headers=headers)
Page = urlopen(req).read() 
Soup = BeautifulSoup(Page,'html.parser')
```

```
Name  = Soup.find('h1', attrs={'class': 'mb-md-0 align-items-start mx-0 my-0'}).text
print(Name)
```

```
PriceMin = Soup.find('div', attrs={'class': 'text-info'}).text
print(PriceMin)
```

```
# ถ้าจะให้ cover กรณีที่ High price error ต้องใช้ try except เข้ามาช่วย
print('Scraping High Price')

try:
    PriceHigh = Soup.find('div', attrs={'class': 'col-sm-9 col-8'}).text
except AttributeError:
    PriceHigh = ''
    
print(PriceHigh)
```

```
## Location
LocationUrl = Soup.find('div', attrs = {'style':'cursor:pointer'}).find('img').get('src')
LocationUrl
```

```
# Image URL
ImageUrl = Soup.find('div', attrs = {'class':'bn-media-cover order-md-2'}).find('img').get('src')
ImageUrl
```

```
def func_initExcel(wb, ws, XlsFile):
  ws['A'+str(1)] = 'Index'
  ws['B'+str(1)] = 'Name'
  ws['C'+str(1)] = 'PriceMin'
  ws['D'+str(1)] = 'HighPrice'
  ws['E'+str(1)] = 'Latitude'
  ws['F'+str(1)] = 'Longitude'
  ws['G'+str(1)] = 'ImageUrl'
  
  wb.save(XlsFile)
```

```
def func_PrintExcel(wb, ws, Var1, Var2, Var3, Var4,Var5,Var6,Var7, Line, XlsFile):
  Line=Line+1
  ws['A'+str(Line)] = Var1
  ws['B'+str(Line)] = Var2
  ws['C'+str(Line)] = Var3
  ws['D'+str(Line)] = Var4
  ws['E'+str(Line)] = Var5
  ws['F'+str(Line)] = Var6
  ws['G'+str(Line)] = Var7
  
  wb.save(XlsFile)
```

```
QuotePage_List = ['https://baania.com/en/project/%E0%B8%9E%E0%B8%B2%E0%B8%A3%E0%B9%8C%E0%B8%84%E0%B9%80%E0%B8%A7%E0%B8%A2%E0%B9%8C-%E0%B9%81%E0%B8%AD%E0%B8%97%E0%B8%AD%E0%B8%B5%E0%B8%8B?s=search&pid=BKK2578&province=3781&district=3936&ptype=1',
'https://baania.com/en/project/%E0%B9%80%E0%B8%99%E0%B9%80%E0%B8%9A%E0%B8%AD%E0%B8%A3%E0%B9%8C%E0%B9%82%E0%B8%AE%E0%B8%A1-%E0%B8%A7%E0%B8%B1%E0%B8%8A%E0%B8%A3%E0%B8%9E%E0%B8%A5?s=search&pid=BKK1914&province=3781&district=3981&ptype=1,7',
'https://baania.com/en/project/%E0%B8%84%E0%B8%B2%E0%B8%8B%E0%B9%88%E0%B8%B2-%E0%B8%A7%E0%B8%B4%E0%B8%A5%E0%B8%A5%E0%B9%8C-%E0%B8%A3%E0%B8%B2%E0%B8%A1%E0%B8%AD%E0%B8%B4%E0%B8%99%E0%B8%97%E0%B8%A3%E0%B8%B2%E0%B8%A7%E0%B8%87%E0%B9%81%E0%B8%AB%E0%B8%A7%E0%B8%99-2?s=search&pid=BKK0001&province=3781&district=3862&ptype=1',              
'https://baania.com/en/project/%E0%B9%80%E0%B8%9E%E0%B8%AD%E0%B8%A3%E0%B9%8C%E0%B9%80%E0%B8%9F%E0%B8%84-%E0%B9%80%E0%B8%9E%E0%B8%A5%E0%B8%AA-%E0%B8%9E%E0%B8%A3%E0%B8%B0%E0%B8%A3%E0%B8%B2%E0%B8%A1-9-%E0%B8%81%E0%B8%A3%E0%B8%B8%E0%B8%87%E0%B9%80%E0%B8%97%E0%B8%9E%E0%B8%81%E0%B8%A3%E0%B8%B5%E0%B8%91%E0%B8%B2?s=search&pid=BKK0968&province=3781&district=3950&ptype=1',
'https://baania.com/en/project/%E0%B8%A5%E0%B8%B1%E0%B8%94%E0%B8%94%E0%B8%B2%E0%B8%A3%E0%B8%A1%E0%B8%A2%E0%B9%8C-%E0%B8%AD%E0%B8%B4%E0%B8%A5%E0%B8%B4%E0%B9%81%E0%B8%81%E0%B8%99%E0%B8%8B%E0%B9%8C-%E0%B8%A7%E0%B8%87%E0%B9%81%E0%B8%AB%E0%B8%A7%E0%B8%99-%E0%B8%AA%E0%B8%B2%E0%B8%97%E0%B8%A3?s=search&pid=BKK0391&province=3781&district=3869&ptype=1']
i=1
wb = openpyxl.Workbook()
ws = wb.active
fileName = 'House_Scraping.xlsx'
func_initExcel(wb,ws,fileName)

for QuotePage in QuotePage_List:
  req = Request(url=QuotePage, headers=headers)
  Page = urlopen(req).read() 
  Soup = BeautifulSoup(Page,'html.parser')
  print('--- House Number '+ str(i) +' ---')
  #Name
  Name  = Soup.find('h1', attrs={'class': 'mb-md-0 align-items-start mx-0 my-0'}).text
  print(Name)
  #PriceMin
  PriceMin = Soup.find('div', attrs={'class': 'text-info'}).text
  print(PriceMin)
  #PriceHigh
  print('Scraping High Price')

  try:
    PriceHigh = Soup.find('div', attrs={'class': 'col-sm-9 col-8'}).text
  except AttributeError:
    PriceHigh = ''
    
  print(PriceHigh)
  ## Location
  print('Perform location scraping')
  LocationUrl = Soup.find('div', attrs = {'style':'cursor:pointer'}).find('img').get('src')
  print(LocationUrl)
  # Image URL
  print('Perform Image scraping')
  ImageUrl = Soup.find('div', attrs = {'class':'bn-media-cover order-md-2'}).find('img').get('src')
  print(ImageUrl)
  func_PrintExcel(wb, ws, i, Name, PriceMin, PriceHigh, LocationUrl, ImageUrl, i, fileName)
  i=i+1
  time.sleep(2)
wb.close()
```
