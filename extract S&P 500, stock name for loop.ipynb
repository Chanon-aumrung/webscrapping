```
# Library for web sracping
from bs4 import BeautifulSoup
from urllib.request import urlopen, Request
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.3'}

# Library for MSExcel
import openpyxl
from google.colab import files

#Time Library
import time
```

```
def func_PrintExcel(wb, ws, Var1, Var2, Var3, Var4, Line, XlsFile):
  ws['A'+str(Line)] = Var1
  ws['B'+str(Line)] = Var2
  ws['C'+str(Line)] = Var3
  ws['D'+str(Line)] = Var4
  
  wb.save(XlsFile)
```

```
def func_initExcel(wb, ws, XlsFile):
  ws['A'+str(1)] = 'Name'
  ws['B'+str(1)] = 'Price'
  ws['C'+str(1)] = 'Change Price'
  ws['D'+str(1)] = 'Change Percent'
  
  wb.save(XlsFile)
```

```
def func_ScrapInvesting(url):
  req = Request(url=url, headers=headers) 
  page = urlopen(req).read() 
  soup = BeautifulSoup(page,'html.parser')
  
  name_box = soup.find('h1', attrs={'class': 'float_lang_base_1 relativeAttr'}) # Take out div of name and get its value
  name = name_box.text.strip()
  
  price_box = soup.find('span', attrs={'id': 'last_last'}) 
  price = price_box.text.strip()
    
  change_box = soup.find('div', attrs={'class': 'top bold inlineblock'})
  change = change_box.text
  
  #print("----- change -------")
  #print(change)
  #print("-----------------------")
  
  change_val  = change.split()[1]
  change_pct = change.split()[2]
  
  print('Name = ' + name)
  print('Price = ' + price)
  print('Change Val  = ' + change_val)
  print('Change Pct  = ' + change_pct)

  return(name,price,change_val,change_pct)
```

```
url_list = ["https://th.investing.com/indices/us-spx-500",
            "https://th.investing.com/equities/pepsico",
            "https://th.investing.com/equities/exxon-mobil"]
```

```
i = 1

wb = openpyxl.Workbook()
ws = wb.active
fileName = 'Loop_export_stock.xlsx'
func_initExcel(wb,ws,fileName)

for url in url_list :
  print('--- Interation Number #' + str(i) + '---')
  
  (name,price,change_val,change_pct) = func_ScrapInvesting(url)
  
  func_PrintExcel(wb, ws, name, price, change_val, change_pct, i, fileName)
  i=i+1
  
  time.sleep(2)
  
wb.close()
print('Program Complete')
```
